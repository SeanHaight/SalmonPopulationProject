```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
SSTs = pd.read_csv("/home/fenton/Documents/Coding/Erdos Project/Salmon Population Project/Data/SSTs")
Sals = pd.read_excel("/home/fenton/Documents/Coding/Erdos Project/Salmon Population Project/Data/ZooPlanktonPerryData.xlsx", 5)
Flow = pd.read_csv("/home/fenton/Documents/Coding/Erdos Project/Salmon Population Project/Data/FraserRiverFlowAtHope-annual_extremes_20231107T1938.csv")
```

```python
#For each year I want a list with the SSTs for the months in ascending order. 
SSTsList = []
for year in SSTs["Year"].unique():
    SSTsList.append([year] + SSTs[SSTs["Year"] == year]["SST"].tolist())

newSSTs = pd.DataFrame(SSTsList, columns=["Year", "Jan.", "Feb.", "Mar.", "Apr.", "May", "Jun.", "Jul.", "Aug", "Sep.", "Oct.", "Nov.", "Dec."])
```

```python
for month in newSSTs.columns[1:]:
    plt.scatter(newSSTs[month][:-5], Sals["Big Qualicum Coho"])
    plt.show()
    plt.close()
```

```python
ModernFlow = Flow[Flow["Year"] >= 1990]
P1Flow = ModernFlow[Flow["PARAM"]==1]
```
