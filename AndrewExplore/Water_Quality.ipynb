```python
Fraser_qual = Fraser_Qual.rename(columns={'DATE_TIME_HEURE':'Time', 'VALUE_VALEUR':'Value', 'UNIT_UNITÉ':'Units', 'VARIABLE':'Variable'})
```

```python
Fraser_qual = Fraser_qual.drop(columns=list(set([col for col in Fraser_qual.columns]) - set(['Time','Value','Units','Variable'])))
Fraser_qual
```

```python
Fraser_qual['Time'] = Fraser_qual['Time'].dt.year
```

```python
var1 = Fraser_qual['Variable'].unique()[0]
var1
```

```python
var1_df = Fraser_qual.loc[Fraser_qual['Variable']==var1]
var1_df
```

```python
#var1_df.groupby('Time').mean(numeric_only=True).reset_index().rename(columns={'Value':var1})
var1_avg = var1_df.groupby('Time').mean(numeric_only=True).rename(columns={'Value':var1})
var1_avg
```

```python
var2 = Fraser_qual['Variable'].unique()[1]
var2_df = Fraser_qual.loc[Fraser_qual['Variable']==var2]
var2_avg = var2_df.groupby('Time').mean(numeric_only=True).rename(columns={'Value':var2})
var2_avg
```

```python
Fraser_water_qual = pd.concat([var1_avg,var2_avg],axis=1)
Fraser_water_qual
```

```python
for i in range(2,len(Fraser_qual['Variable'].unique())):
    var = Fraser_qual['Variable'].unique()[i]
    var_df = Fraser_qual.loc[Fraser_qual['Variable']==var]
    var_avg = var_df.groupby('Time').mean(numeric_only=True).rename(columns={'Value':var})
    Fraser_water_qual = pd.concat([Fraser_water_qual,var_avg],axis=1)

Fraser_water_qual
```
